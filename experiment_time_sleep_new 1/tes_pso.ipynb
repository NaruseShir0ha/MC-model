```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#从第一行开始读
data=pd.read_excel('excel_folder/pso_result.xlsx',header=0)
data
```

```python
data=data.values.reshape(16,24)
data
```

```python
#将每一组的数字做散点图,横轴未组别，纵轴为适应度值,减小散点大小，将data值log10
plt.figure(figsize=(15,10))
plt.xlabel('group')
plt.ylabel('lg(fitness)')
plt.title('lg(fitness) of each group')
plt.xticks(range(15))
plt.yticks(np.arange(-0.5,2.5,0.1))
plt.ylim(-0.5,2.5)
plt.grid()
for i in range(15):
    plt.scatter([i]*24,np.log10(data[i]),s=10)
plt.savefig('png_folder/fitness of each group.png')
plt.show()
```
